```python
import pandas as pd

data = pd.read_csv('breast-cancer.csv')
data.head()
```

```python
data.info()
```

```python
df = data.copy()
```

```python
import numpy as np

def repair_colum(x: str) -> float:
    numeros = x.split('-')
    nuevos_numeros = []
    for n in numeros:
        nuevos_numeros.append(float(n))
    return np.mean(nuevos_numeros)
```

```python
df['age'].apply(repair_colum)
```

```python
df.head()
```

```python
df['age'] = df['age'].apply(repair_colum)

df.head()
```

```python
df = data.copy()

colums2repair = ['age', 'tumor-size', 'inv-nodes']
for col in colums2repair:
    df[col] = df[col].apply(repair_colum)
    
df.head()
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)

df_trans = pca.fit_transform(df[ colums2repair + ['deg-malig']])
```

```python
from sklearn.cluster import KMeans
```

```python
clf = KMeans(n_clusters=2)

clf.fit(df_trans)

pred = clf.predict(df_trans)
```

```python
pred
```

```python
df_trans
```

```python
df[df.columns[-2]].unique()
```

```python
def str2clas(x):
    if 'recurrence-events' == x:
        return 0
    return 1
df['clases_numericas'] = df[df.columns[-1]].apply(str2clas)

df.head()
```

```python
df['clases_pred'] = pred
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(12, 7))

ax = fig.add_subplot()
ax.scatter(x=clf.cluster_centers_[:, 0], y=clf.cluster_centers_[:, 1], c='Red', label='Clusters')
ax.scatter(x=df_trans[:, 0], y=df_trans[:, 1], c=df['clases_pred'].values, alpha=.3, label='PCA preds')
ax.scatter(x=df_trans[:, 0], y=df_trans[:, 1], c=df['clases_numericas'].values, alpha=.3, label='clf real')
ax.set_title('Titulo del grafico')
ax.set_xlabel('PCA_1')
ax.set_ylabel('PCA_1')
ax.grid()
ax.legend()
plt.show()
```
